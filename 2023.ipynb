```python
import cv2 as cv
import os
```

```python
img_path=os.path.join('.','img.jpeg')

img=cv.imread(img_path)

cv.imwrite(os.path.join('.','img_out.jpeg'),img)
```

```python
video_path=os.path.join('.','output.avi')

video=cv.VideoCapture(video_path)

ret=True

while ret:
    ret,frame=video.read()
    if ret:
        cv.imshow('frame',frame)
        if cv.waitKey(20) & 0xFF == ord('q'):
            break

video.release()
cv.destroyAllWindows()
```

```python
cam=cv.VideoCapture(0)


while True:
    ret, frame=cam.read()
    cv.imshow('frame',frame)
    if cv.waitKey(40) & 0xFF == ord('q'):
        break
cam.release()
cv.destroyAllWindows()
```

```python
pic=os.path.join('.','img.jpeg')

img=cv.imread(pic)
resize=cv.resize(img, (200,200))

print(img.shape)
print(img.shape)
cv.imshow('pic',img)
cv.imshow('resize',resize)

cv.waitKey(0)
cv.destroyAllWindows()
```

```python
img_path=os.path.join('.','img.jpeg')

img=cv.imread(img_path)

cv.imshow('img',img)

crop_img=img[80:145,70:160]
cv.imshow('crop_img',crop_img)
print(img.shape)
cv.waitKey()
cv.destroyAllWindows()
```

```python

COLOR=cv.cvtColor(img,cv.COLOR_BGR2HSV)

cv.imshow('img',img)
cv.imshow('color',COLOR)
cv.waitKey(0)
cv.destroyAllWindows()
```
